```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
```

# County FIPS data aquired from 
https://www.census.gov/geographies/reference-files/2017/demo/popest/2017-fips.html

```python
file="MNcounty-geocodes-v2017.csv"
fips_df=pd.read_csv(file)
fips_df.head()
```

# Select out Minnesota counties only

```python
mn_fips=fips_df[fips_df['State Code (FIPS)'] == 27]
mn_fips=mn_fips[mn_fips['County Subdivision Code (FIPS)'] == 0]
mn_fips=mn_fips[mn_fips['Place Code (FIPS)'] == 0]
mn_fips=mn_fips[mn_fips['County Code (FIPS)'] != 0]
mn_fips= mn_fips[["State Code (FIPS)","County Code (FIPS)","Area Name (including legal/statistical area description)"]]
mn_fips.head()
```

```python
mn_fips=mn_fips.rename(columns={"State Code (FIPS)":"statefips","County Code (FIPS)":"countyfips","Area Name (including legal/statistical area description)":"county"})
mn_fips.head()
```

# remove 'County' from name and make lower case

```python
new_county=[]
for row in mn_fips['county']:
    if row[0] != 'C':
        cnty=row.split('C')[0].lower() 
    elif row == 'Crow Wing County':
        cnty = 'crow wing'
    else:
        cnty=row.split(' ')[0].lower()
    print(cnty)
    new_county.append(cnty)
mn_fips['county']=new_county
mn_fips.head()
```

# Tableau requires state and country also

```python
mn_fips["state"]= 'minnesota'
mn_fips["country"] = 'united states'
mn_fips.head()
```

```python
mn_fips.to_csv('mnCountyFIPS.csv', index=False,header=True,encoding='utf-8')
```
