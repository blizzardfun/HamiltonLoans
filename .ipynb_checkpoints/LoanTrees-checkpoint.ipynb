```python
from sklearn import tree
import pandas as pd
```

```python
USB= 504713  # US Bank
Wells=451965  #Wells Fargo
Quicken=7197000003        #Quicken Loan

loans_raw = pd.read_csv('AllData2017.csv')
loans_df = loans_raw.loc[(loans_raw['respondent id'] == Wells) | (loans_raw['respondent id'] == USB) | (loans_raw['respondent id'] == Quicken), :]
loans_df.head()
```

```python
target = loans_df["action modified"]
target_names = ["declined", "approved"]
```

```python
data = loans_df.drop(["action modified","respondent id", "owner occ", "state", "county", "loan type", "census tract", "hoepa status", "Index", "agency code", "income", "msa/md", "pre approval", "action taken", "denial reason 1", "denial reason 2", "rate of spread", "denial reason 3", "type of purchaser"],axis=1)
feature_names = data.columns
data.head()
```

```python
# Split the data using train_test_split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target)
```

```python
# Create, fit, and score a Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=50)
rf = rf.fit(X_train, y_train)
```

```python
# Calculate the R2 score for the test data
print(f"Training Data Score: {rf.score(X_train, y_train)}")
print(f"Testing Data Score: {rf.score(X_test, y_test)}")
```

```python
# Random Forests in sklearn will automatically calculate feature importance
importances = rf.feature_importances_
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```
